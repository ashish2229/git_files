```python
import csv
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter

with open('C:/Users/DELL/Desktop/ML/data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    data_counter=Counter()
    
    for row in csv_reader:
        data_counter.update(row['LanguagesWorkedWith'].split(';'))
        

language=[]
popularity=[]

for item in data_counter.most_common(15):
    language.append(item[0])
    popularity.append(item[1])

print(language)
print(popularity)

plt.barh(language,popularity) #for printing horizontal bars
plt.ylabel("programming language")
plt.xlabel("No of users")
plt.title("Top 15 languages used")

plt.show()

# now if we want to print the bar chart in ascending order, we have to use reverse

language.reverse()
popularity.reverse()

plt.barh(language,popularity) #for printing horizontal bars
plt.ylabel("programming language")
plt.xlabel("No of users")
plt.title("Top 15 languages used")

plt.show()
```
